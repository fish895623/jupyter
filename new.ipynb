```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

fuel_econ = pd.read_csv(
    "https://raw.githubusercontent.com/BaekKyunShin/Data-Analyst-Nanodegree/master/Practice-Plot/fuel-econ.csv")
fuel_econ.head()
```

```python
plt.scatter(data=fuel_econ, x='city', y='highway', alpha=1 / 8)
plt.show()
```

```python
sb.regplot(data=fuel_econ,
           x='city',
           y='highway',
           fit_reg=False,
           scatter_kws={'alpha': 0.2})
```
